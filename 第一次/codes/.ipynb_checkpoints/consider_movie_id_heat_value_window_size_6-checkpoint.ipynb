```python
df = pd.read_csv("../first_version/cleaned_features/original_df_2.csv", index_col=0)
```

```python
df.head()
```

```python
df['电影ID'] = original_df['电影ID']
```

```python
def get_heat_values_windows_size_6(data):
    
    new_now_heat_value = []
    
    if data.shape[0] < 6:
        new_now_heat_value = data['当前热度值'].values.tolist()
        
    else:
        # 此处-5是因为dataframe格式，尽管只剩最后2行，写成.iloc[-6:, :]也不会报错，但是只计算最后两行的值，结果就不对了。
        for i in range(0, data.shape[0]-5): 
            new_now_heat_value.append(data.iloc[i:i + 6, :]['当前热度值'].mean())
        for i in range(5):
            new_now_heat_value.insert(i, 0)
            
    return new_now_heat_value
```

```python
del df['电影ID']
```

```python
all_new_now_heat_value = []
for movie_id in df['电影ID'].unique().tolist():
    print("movie id:", movie_id)
    all_new_now_heat_value.extend(get_heat_values_windows_size_6(df[df['电影ID']==movie_id]))
    print("length of new list:",len(all_new_now_heat_value))
```

```python
df['heat_value_window_size_6'] = pd.DataFrame(all_new_now_heat_value)
```

```python
df.to_csv("../first_version/cleaned_features/original_df_update_heat_value_window_size_6_with_movie_id.csv")
```

```python
X = df.values.tolist()
```

```python
## 保留按照movei_id生成的'heat_value_window_size_6'，用最优参数跑

import xgboost as xgb
from xgboost import plot_importance
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

model = xgb.XGBRegressor(max_depth=7, n_estimators=500, gamma=0.027698143119166907, colsample_bytree=0.8957822073838446)
model.fit(X_train, y_train)

# 保存模型
model.save_model("../first_version/model_results/y_new_add_update_heat_value_window_size_6.model")

# 对测试集进行预测
ans = model.predict(X_test)

print("MSE:",metrics.mean_squared_error(y_test, ans))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, ans))) 
```
